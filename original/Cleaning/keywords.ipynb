```python
# remove stop words function
stopwords = ['of','on','all','the','a','for','to','and','to','against','from','in','or']

def noStop(text):
    '''
    this function inputs a text, lowers it, and removes stop words
    '''
    text = text.lower().split()
    for word in stopwords:
        text = [x for x in text if x != word]
    return(' '.join(text))
```

## II. Actions

```python
# assign action
keywords = {}
keywords['ratify'] = ['ratif']
keywords['accede'] = ['acced']
keywords['sign'] = ['sign ']
keywords['implement'] = ['implement']
keywords['establish'] = ['establish']
keywords['continue'] = ['continue']
keywords['take-measures'] = ['take measures']
keywords['harmonize'] = ['harmoniz']
keywords['report'] = ['report','reporting','reports']
keywords['party'] = ['party']
keywords['withdraw'] = ['withdraw']
keywords['cooperate'] = ['cooperat']
keywords['share'] = ['share', 'sharing']

print (keywords.keys())
```

```python
for i in recs:
    i['action'] = []
    for key in keywords.keys():
        for word in keywords[key]:
            if key in i['action']: # skip if it's already in there
                continue
            if word in noStop(i['text']):
                i['action'].append(key)
```

```python
# special rules 

# international assistance
for i in recs:
    if (
        ('assistance ' in i['text'] or 'support ' in i['text']) and
        'international' in i['text']
        ):
        i['action'].append('assistance')
        

# take ____ measures:
for i in recs:
    if 'take-measures' not in i['action']:
        text = i['text'].lower().replace(',','').split()
        if 'take' in text:
            takeIndex = text.index('take')
            if 'measures' in text[takeIndex:takeIndex+10]:
                i['action'].append('take-measures')
        elif 'taking' in text:
            takeIndex = text.index('taking')
            if 'measures' in text[takeIndex:takeIndex+10]:
                i['action'].append('take-measures')
```

```python
#apply
for i in recs:
    i['theme'] = []
    text = noStop(i['text'])
    # get rid of country, so that 'democrat' doesn't catch "democratic republic of congo" etc.
    if "(" in text and ")" in text:
        country = text.split('(')[-1].split(')')[0]
        text = text.replace(country,'')
    for key in keywords.keys():
            for word in keywords[key]:
                if key in i['theme']: # skip if it's already in there
                    continue
                if word in text:
                    i['theme'].append(key)
```

```python
# Special Rules

#"NGO" (capitalization required)
for i in recs:
    if 'civil-so' not in i['theme']:
        if "NGO" in i['text']:
            i['theme'].append('civil-so')
            
# HIV (capitalization required)
for i in recs:
    if 'hiv' not in i['theme']:
        if "AIDS" in i['text']:
            i['theme'].append('hiv')

#"THE poor" (stopword required)
for i in recs:
    if 'poor' not in i['theme']:
        if "the poor" in i['text']:
            i['theme'].append('poor')

# "courts" and "tribunal" but not ICC, ICT, European court etc (lowercase requried)
for i in recs:
    if 'judiciary' not in i['theme']:
        if 'court' in i['text'] or 'tribunal' in i['text']:
            i['theme'].append('judiciary')
    
# in custody (stop words required)
for i in recs:
    if 'prisoners' not in i['theme']:
        if 'in custody' in i['text']:
            i['theme'].append('prisoners')
            
# assembly = civil society (lowercase requried)
for i in recs:
    if 'civil-so' not in i['theme']:
        if 'assembly' in i['text']:
            i['theme'].append('civil-so')
        
# if sex-violence, then women
for i in recs:
    if 'sex-violence' in i['theme']:
        if 'women' not in i['theme']:
            i['theme'].append('women')
```

```python
# exploring

def topThemes(country):
    allThemes = []
    for i in recs:
        if i['to'] == country:
            if i['theme'] != []:
                for theme in i['theme']:
                    allThemes.append(theme)
    return(Counter(allThemes))
```

```python
def getTheme(country,theme):
    allRecs = []
    for i in recs:
        if i['to'] == country:
            if theme in i['theme']:
                allRecs.append(i['text'])
    return(allRecs)
```
